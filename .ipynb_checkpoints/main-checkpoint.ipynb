```python
game_state = { # create a scoreboard dictionary to track the game
    "score": 0, # score starts at 0 - no correct answers yet
    "current_question": 0, # current question starts at 0 - we're on the first question
    "lives": 3, # lives start at 3 - player gets 3 chances
    "game_over": False # game over starts as False - the game hasn't ended yet
}    
```

```python
questions = [ # create a list of questions - each question is its own dictionary
    {"question": "In 'The Lord of the Rings: The Return of the King', who kills the Witch-king of Angmar?",
     "options": ["A) Aragorn", "B) Eowyn", "C) Gandalf", "D) Frodo"],
     "answer": "B"}, #each dictionary has three things: "question", "options", "answer"; the list holds all 5 questions together like a deck of cards

    {"question": "In 'Pulp Fiction', what dance contest do Vincent and Mia enter?",
     "options": ["A) The Twist", "B) The Tango", "C) The Jive", "D) The Waltz"],
     "answer": "A"},

    {"question": "In 'Interstellar', what is the name of the robot companion that accompanies Cooper?",
     "options": ["A) R2-D2", "B) HAL", "C) WALL-E", "D) TARS"],
     "answer": "D"},

    {"question": "In 'Gladiator', what is the name of Maximus's gladiator trainer?",
     "options": ["A) Cassius", "B) Tigris", "C) Proximo", "D) Quintus"],
     "answer": "C"},

    {"question": "In 'Star Wars: Return of the Jedi', who throws Emperor Palpatine into the reactor shaft?",
     "options": ["A) Luke Skywalker", "B) Darth Vader", "C) Han Solo", "D) Yoda"],
     "answer": "B"},
]    
```

```python
def display_question(question): # function to show one question to the player
    print(question["question"]) # print the question text from the dict
    for option in question["options"]: # loop through the 4 options
        print(option) # print each option on its own line
```

```python
def process_turn(question, game_state): # function to play one round
    display_question(question) # show the question first
    valid_answer = False # gate starts closed - no valid answer yet
    while valid_answer == False: # keep asking until we get a valid answer
        answer = input("Your answer: ").upper() # input, convert to uppercase so a and A both work
        if answer == "A" or answer == "B" or answer == "C" or answer == "D": # check if input is valid
            valid_answer = True # gate opens - valid answer recieved, exit loop
        else: 
            print("Invalid, please enter A, B, C or D") # reject invalid input, loop again
    
    if answer == question["answer"]: # compare player's answer to correct answer
        print("Correct!") # player got it right
        game_state["score"] += 1 # add 1 to score on the scoreboard
    else:
        print(f"Wrong, the correct answer was {question['answer']}.") # show correct answer
        game_state["lives"] -= 1 # add 1  to score on the scoreboard
```

```python
def start_game(): # main function that runs the whole game
    game_state["score"] = 0 # reset score to zero
    game_state["current_question"] = 0 # reset to the first question
    game_state["lives"] = 3 # reset lives to 3
    game_state["game_over"] = False # reset game over flag

    print("Welcome to the Film Quiz!") # greet the player

    while not game_state["game_over"]: # keep looping until game ends
        question = questions[game_state["current_question"]] # grab the current question from the list using index
        process_turn(question, game_state) # play one round with that question
        game_state["current_question"] += 1 # move to the next question

        if game_state["lives"] <=0: # check lose condition - no lives left
            game_state["game_over"] = True # end the game
            print(f"Game over, you ran out of lives. Final score: {game_state['score']}/{len(questions)}") # show final score
        elif game_state["current_question"] >= len(questions): # check win condition - all questions answered
            game_state["game_over"] = True # end the game
            print(f"You survived the quiz! Final score: {game_state['score']}/{len(questions)}") # show final score
```

```python
start_game()
```
